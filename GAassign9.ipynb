```python
#Importing necessary libraries
# Seaborn and Matplotlib are imported for data visualization
import seaborn as sns
#matplotlib is a library and pyplot is the submodule used for plotting graphs
from matplotlib import pyplot as plt #or import matplotlib.pyplot as plt
```

```python
#Load and display the dataset
# Data from titanic dataset is loaded in df variable.To acheive this load_dataset 
#function is used from seaborn library
df = sns.load_dataset('titanic')
# df.head() displays first few rows of the dataset, according to the parameter 
#passed. By default the head function displays 5 rows.
# It is used to get an initial understanding of dataset structure
df.head()
```

```python
#Boxplot for Distribution of Age by Gender
# Printing column names to identify the correct columns for 
#the boxplot's x and y axes
print(df.columns)
```

```python
# figure is a function in matplotlib.pyplot which allows us to 
#create a new figure it takes various parameters like
#1. figsize(float, float): These parameter are the width, height in
#inches. 2. facecolor: This parameter is the background color.
plt.figure(figsize=(10,7))
# title function from matplotlib.pyplot takes a 
#string as parameter. This string specifies the title of the plot
plt.title('Distribution of Age by Gender')
#Sets the label of X axis as Gender
plt.xlabel('Gender')
#Sets the label of Y axis as Age
plt.ylabel('Age')
# Using Seaborn's boxplot to create a box plot, which shows the distribution 
#of ages (y='age') grouped by gender (x='sex') 
# boxplot takes following parameter : x = feature of dataset, 
#y = feature of dataset, data = dataframe or full dataset
sns.boxplot(x='sex', y='age', data=df)
# plt.show() is required when running in non-interactive environments. 
#In case of non-interactive environments the plot maybe saved in memory but
#won't be displayed. Also this gives us power to control when the plot is displayed. 
#Helps in displaying changes made in the plot after its creation
plt.show()
```

```python
#Boxplot for Distribution of Age by Gender and Survial Status
# Create a new figure with its width as 10 and height as 8 inches
plt.figure(figsize=(10,7))
#Set the title of the plot as 
#Distribution of Age by Gender and Survial Status
plt.title('Distribution of Age by Gender and Survial Status')
#Sets the label of X axis as Gender
plt.xlabel('Gender')
#Sets the label of Y axis as Age
plt.ylabel('Age')
# Using Seaborn's boxplot to create a box plot, which shows the 
#distribution of ages (y='age') grouped by gender (x='sex') and futher 
# differentiated by survival status (hue='survived'). The hue parameter 
#adds another dimension to the plot, allowing us to see the
# the survival status within each gender group
# boxplot takes an additional parameter : hue = feature of dataset
sns.boxplot(x='sex', y='age', data=df, hue='survived')
# plt.show() is required when running in non-interactive environments. 
#In case of non-interactive environments the plot maybe saved in memory but
#won't be displayed. Also this gives us power to control when the plot is displayed. 
#Helps in displaying changes made in the plot after its creation
plt.show()
```

```python
# Create a new figure with its width as 10 and height as 8 inches
plt.figure(figsize=(10,20))
#Set the title of the plot as Distribution of fare by passenger class
plt.title('Distribution of Fare by Passenger Class and Survival Status')
#Sets the label of X axis as Passenger Class
plt.xlabel('Passenger Class')
#Sets the label of Y axis as Fare
plt.ylabel('Fare')
# Using Seaborn's boxplot to create a box plot, which shows the 
#distribution of fares (y='fare') grouped by passenger class (x='pclass') 
# The hue parameter adds another dimension to the plot, allowing 
#us to see the survival status within each gender group
# using boxplot function from seaborn library. This plot helps 
#visualize the variation in fares across different passenger classes.
sns.boxplot(x='pclass', y='fare', data=df, hue='survived')
# plt.show() is required when running in non-interactive environments. 
#In case of non-interactive environments the plot maybe saved in memory but
#won't be displayed. Also this gives us power to control when the plot is displayed. 
#Helps in displaying changes made in the plot after its creation
plt.show()
```
